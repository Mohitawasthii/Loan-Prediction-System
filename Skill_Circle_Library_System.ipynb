```python
def book_info(book_no):
    if book_no == 1:
        print("Title : Superman: Man of Steel")
        print("Author Name: DC Comics")
        print("Description: A comic book about Superman.")
        print("Book Price: ₹350")
        print("Launch Date: April 10, 2023")

    elif book_no == 2:
        print("Title : Iron Man: The Invincible")
        print("Author Name: Marvel Comics")
        print("Description: A comic book about Iron Man.")
        print("Book Price: ₹400")
        print("Launch Date: May 15, 2023")

    elif book_no == 3:
        print("Title : Chacha Chaudhary and Sabu")
        print("Author Name: Pran Kumar Sharma")
        print("Description: A wise and clever old man.")
        print("Book Price: ₹150")
        print("Launch Date: June 20, 2023")

    elif book_no == 4:
        print("Title : Champak Stories Collection")
        print("Author Name: Champak Magazine")
        print("Book Price: ₹699")

    elif book_no == 5:
        print("Title : Chandamama Tales")
        print("Author Name: Chandamama Magazine")
        print("Book Price: ₹499")

    else:
        print("Invalid book number")


def info():
    try:
        choice = int(input("Enter 1 to continue or 2 to exit: "))

        if choice == 1:
            try:
                book_no = int(input("Enter book number (1-5): "))
                book_info(book_no)
            except ValueError:
                print("❌ Please enter a valid number for book")

            info()

        elif choice == 2:
            print("Thank you for using Skill Circle Library System")

        else:
            print("Invalid input")
            info()

    except ValueError:
        print("❌ Please enter only numbers (1 or 2)")
        info()

try:
    book_no = int(input("Enter book number (1-5): "))
    book_info(book_no)
except ValueError:
    print("❌ Invalid input! Please enter a number.")

info()
```
